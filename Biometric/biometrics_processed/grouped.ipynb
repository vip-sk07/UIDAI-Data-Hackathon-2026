```python
import pandas as pd
df = pd.read_csv("E:\\DATA ANALYTICS\\biometrics_cleaned.csv")
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
monthly_data = df.groupby(['year', 'month'])[
    ['bio_age_5_17', 'bio_age_17_']
].sum().reset_index()
monthly_data.to_csv(
    r"E:\DATA ANALYTICS\biometrics_monthly_aggregated.csv",
    index=False
)
print("Monthly aggregation completed successfully")
print(monthly_data.head())
print("\nShape of aggregated data:", monthly_data.shape)
```
