```python
import pandas as pd
import numpy as np

# Load processed enrolment data
file_path = r"D:\Academics\Hackathon\UIDAI Data HAckathon-2026\enrolment\enrolement_processed.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df.info()
df.describe()
df.isna().sum()
```

```python
#Key Metrics
# Total population covered (normalized)
total_population = df['total_population_norm'].sum()

# Age-wise enrolment
age_0_5 = df['age_0_5_norm'].sum()
age_5_17 = df['age_5_17_norm'].sum()
age_18_plus = df['age_18_greater_norm'].sum()

# Ratios (average)
avg_child_ratio = df['child_ratio'].mean()
avg_adult_ratio = df['adult_ratio'].mean()

print("Total Population (Normalized):", total_population)
print("Age 0–5 Enrolment:", age_0_5)
print("Age 5–17 Enrolment:", age_5_17)
print("Age 18+ Enrolment:", age_18_plus)
print("Average Child Ratio:", round(avg_child_ratio, 2))
print("Average Adult Ratio:", round(avg_adult_ratio, 2))
df['enrolment_pressure'].value_counts()

```

```python
#Cluster-wise Enrolment
cluster_dist = df.groupby('enrolment_cluster')['total_population_norm'].sum()
cluster_dist
```

```python
## Enrolment Cluster Interpretation

- Enrolment clusters group districts with similar enrolment behavior
- Higher cluster values represent regions with higher enrolment coverage
- Cluster with maximum population indicates high enrolment density areas
- Low clusters may require targeted enrolment drives
```
